## FORMULA SHANNON ENTROPIA:
$$\text{shannon} = -\displaystyle\sum_{k=1}^n Pk \cdot $log{_{logbase}}({Pk})$ $$ DOVE $Pk = frequenza/dataSize$

Funzione matematica che intuitivamente corrisponde alla quantità di informazioni contenute o fornite da una fonte di informazione.
Dal punto di vista di un ricevente, più diverse sono le informazioni che la sorgente emette, maggiore è l'entropia (o l'incertezza su ciò che la sorgente emette). Quindi, se una sorgente invia sempre lo stesso simbolo, ad esempio la lettera 'a', la sua entropia è zero, cioè minima. Infatti, un ricevitore che conosce solo le statistiche di trasmissione della sorgente ha la certezza che il simbolo successivo sarà una "a". D'altra parte, se la fonte invia una "a" metà delle volte e una "b" l'altra metà, il destinatario è incerto sulla lettera successiva da ricevere. L'entropia della sorgente in questo caso è quindi diversa da zero (positiva) e rappresenta quantitativamente l'incertezza che regna sull'informazione proveniente dalla sorgente. L'entropia indica quindi la quantità di informazioni necessarie affinché il ricevitore determini in modo inequivocabile ciò che la sorgente ha trasmesso. Più informazioni il ricevitore riceve sul messaggio trasmesso, più aumenta l'entropia (incertezza) rispetto a questo messaggio. In particolare, più la fonte è ridondante, meno informazioni contiene. In assenza di vincoli particolari, l'entropia è massima per una sorgente di cui tutti i simboli sono ugualmente probabili.

```python
def shannon(data):
    LOG_BASE = 2    #perché l'entropia ha quindi unità di bit / simboli

   # We determine the frequency of each byte
   # in the dataset and if this frequency is not null we use it for the
   # entropy calculation
    dataSize = len(data)    #prendiamo lunghezza del dato
    ent = 0.0               #inizializzo entropia
    freq={}                 #dizionario corrispondente alla frequenza dei char

    #alg che calcola la frequenza dei char
    for c in data:
        if c in freq:
            freq[c] += 1
        else:
            freq[c] = 1

   # to determine if each possible value of a byte is in the list
    for key in freq.keys():
        f = float(freq[key])/dataSize
        if f > 0: # to avoid an error for log(0)
            ent = ent + f * math.log(f, LOG_BASE)
    return -ent

def pre_entropy(payload):
    characters=[]
    for i in payload:
            characters.append(i)
    return shannon(characters)
            
```
